```python
'''
###9498
시험 점수를 입력받아 90 ~ 100점은 A, 80 ~ 89점은 B, 70 ~ 79점은 C, 60 ~ 69점은 D, 나머지 점수는 F를 출력하는 프로그램을 작성하시오.
'''

A2 = int(input(""))

if 90 <= A2 <= 100:
    print("A")
elif 80 <= A2:
    print("B")
elif 70 <= A2:
    print("C")
elif 60 <= A2:
    print("D")
else:
    print("F")
    
    
### 다른풀이
A2 = int(input("점수를 입력하세요: "))

# 등급과 해당 점수 범위를 튜플로 정의 (등급, 최소 점수)
grades = [("A", 90), ("B", 80), ("C", 70), ("D", 60)]

# 점수에 맞는 등급 찾기
for grade, min_score in grades:
    if A2 >= min_score:
        print(grade)
        break
else:
    print("F")
```
